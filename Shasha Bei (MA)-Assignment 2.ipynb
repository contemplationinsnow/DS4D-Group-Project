# Explore,Reflect and Hypotheses 

```python
# In visualisation chart 1, I will examine the percentage of deceased people in different demographic categories for the 5 broad underlying causes of death. Explore whether deprivation, age, ethnic group and sex have an effect on the mortality rate of patients with the 5 underlying causes of death.
demographic_data=demographic_data.loc[demographic_data['Narrow underlying cause of death']=='All']
causeList = demographic_data['Broad underlying cause of death'].unique().tolist()
causeList.remove("All")
print(causeList)
DemographicList = demographic_data['Demographic category'].unique().tolist()
DemographicList.remove("All")
# Visualisation
fig, axs = plt.subplots(len(causeList), len(DemographicList),figsize=(20, 20))
for c in causeList:
    cause_data = demographic_data.loc[demographic_data['Broad underlying cause of death'] == c]
    for d in DemographicList:
        dfDemographic = cause_data.loc[cause_data['Demographic category'] == d]
        # Select the column of data to be plotted
        labels = dfDemographic['Demographic group']
        axs[causeList.index(c), DemographicList.index(d)].pie(dfDemographic['Number of deaths as percentage deaths of deaths with this underlying cause of death'],  labels=labels, pctdistance=0.9)
        axs[causeList.index(c), DemographicList.index(d)].axis('equal')  # 
        axs[causeList.index(c), DemographicList.index(d)].set_title(d)
plt.show()
```

Findings:
1In all underlying causes of death, there is little difference in the proportion of deceased people in different poverty levels.
2In all causes of death, the proportion of White deaths is much greater than for other races. This should be related to the proportion of the native British population, the White British population accounted for approximately 80% of the total population.
3In the emographic category of gender, women accounted for a much larger proportion of deaths due to dementia than men, while the rest of the causes of death were more or less the same for both sexes.
4Among those who died of cancer, the proportion of those younger than 75 years old was the highest among the five basic causes of death.
5Among the 3 basic causes of death, namely cardiovascular, dementia, and respiratory, the number of 85+ deceased is much higher.

Reflection and hypothesis:
1In response to the fact that there is a higher percentage of females than males among those who died from dementia, I propose the hypothesis that females are more likely to develop dementia than males.
2The reason for this hypothesis is that in all of the other 4 basic causes of death in the chart, the percentage of male decedents is slightly higher than the percentage of female decedents, and only the percentage of female decedents is much higher than the percentage of male decedents among those who died due to dementia.

Looking for studies that have addressed the prevalence of dementia in females versus males. Analysing the available datasets through statistical methods and considering whether factors such as age, lifestyle, genetics and comorbidities have an impact on mortality rates of people with dementia to see if there is indeed a correlation between gender and the development of dementia.
```python
# In visualisation chart 2, I will examine the relationship between broad underlying causes of death and the average emergency admissions in the last 3 months of life
emergency_data = emergency_data.loc[emergency_data['Narrow underlying cause of death'] == 'All'][1:]
# different place
causeList = emergency_data['Broad underlying cause of death'].unique().tolist()
emergency_data['Number of people with 1 or 2 emergency hospital admissions in last 3 month of life']=emergency_data['Number of people with at least emergency hospital admission in last 3 month of life']-emergency_data['Number of people with at least 3 emergency hospital admissions in last 3 month of life']
data1=emergency_data['Number of people with 1 or 2 emergency hospital admissions in last 3 month of life']
data2=emergency_data['Number of people with at least 3 emergency hospital admissions in last 3 month of life']

# Set the graphic size
plt.figure(figsize=(10, 6))
# Width of bar
bar_width = 0.35
# Position of the x-axis
x = range(len(causeList))
# Drawing the bars of P1or2
plt.bar(x, data1, bar_width, label='1or2', align='center')
# Drawing P3ormore's Bars
plt.bar([i + bar_width for i in x],data2 , bar_width, label='3orMore', align='center')
# Setting the x-axis labels
plt.xticks([i + bar_width/2 for i in x], causeList)
# Add legend
plt.legend()
# Setting up titles and labels
plt.title('The relationship between Broad underlying cause of death and  Average emergency admissions in last 3 months of life')
plt.xlabel('Broad underlying cause of death')
plt.ylabel('Number of death')
# Display graphics
plt.show()
```

Findings:
1The number of people with cancer was highest among people with 3 or more emergency admissions to hospital.
2People with dementia had the lowest number of people with 1 or 2 emergency admissions and 3 or more. It is possible that dementia is not harmful in itself, but may cause other reasons for emergency admission.
3The highest number of emergency admissions are for other conditions, which may include emergencies such as broken bones and fever, whereas the other conditions in the graph, such as cancer, are chronic long-term conditions.

Reflection and hypothesis:
1In response to the phenomenon that cancer patients have the highest number of emergency admissions to hospital 3 or more, I propose the hypothesis that cancer patients' conditions deteriorate faster in the last 3 months, which can easily lead to other complications such as infections, bleeding, or organ failure that require emergency admissions to hospital for treatment.
2I will test this hypothesis by firstly analysing the medical records and databases of cancer patients to compare their health status, hospital admissions and complications over a longer period of time, especially in the last three months of life. In addition, I will interview patients, carers and healthcare providers to collect qualitative data on the patient's experience in the last months and the changes in the patient's health observed by healthcare professionals, including deterioration and emergencies that lead to hospitalisation.

```python
# In visualisation chart 3, I will examine the relationship between broad underlying causes of death and place of death.
place_data = place_data.loc[place_data['Narrow underlying cause of death'] == 'All']

# Different place
causeList = place_data['Broad underlying cause of death'].unique().tolist()
causeList.remove("All")
PlaceList = place_data['Place of death'].unique().tolist()
plt.figure(figsize=(10, 6))
print(PlaceList)
# Visualisation
x = np.arange(len(PlaceList))
# Setting the x-axis labels
plt.xticks(x, PlaceList)
# Setting the y-axis labels
plt.ylabel('Number of death')
for c in causeList:
    dfVis = place_data.loc[place_data['Broad underlying cause of death'] == c]
    # Create side-by-side line charts
    plt.plot(x, dfVis['Number of death'], marker='o', label=c, linestyle='-')
# Setup legend
plt.legend()
# Setting the title
plt.title('The relationship between Broad underlying cause of death and Place of death')
plt.show()
```

Findings:
1The number of people who die at hospice from all causes except cancer deaths is much smaller than anywhere else. It may be because they are well cared for and comfortable here.
2Except for cancer and dementia patients, the number of deaths at hospital for all other underlying causes is greater than elsewhere.
3The number of cancer patients who die at home is the highest, higher than other places such as hospital, hospice, and care home.
4Most of the dementia patients died in care home, probably most of the families chose to send the dementia patients to the care centre to spend their last time.

Reflection and hypothesis:
1In response to the phenomenon that the largest number of cancer patients die at home, I propose the hypothesis that cancer patients are in the last stage of life, when medical technology can no longer do anything, and what patients need most is emotional support. A home environment gives patients a sense of familiarity and comfort, surrounded by loved ones who provide emotional support in the final stages of life.
2To address this hypothesis, I will test this hypothesis in the following ways: firstly I will interview cancer patients and their families to understand their perspectives and to assess the emotional and psychological needs of patients in the end stage of life. Secondly, I will interview healthcare providers, including social workers, to gain their experiences and insights into the importance of emotional support in end-of-life care.
